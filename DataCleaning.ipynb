# Parsing Date

We can see the datatype of 'date' column is object. So, First lets' parse the date and convert it to dtype 'datetime64'.

```python
data['date_parsed'] = pd.to_datetime(data['date_parsed'],format="%Y-%m-%d")
data['date_parsed']
```

```python
data.head()
```

### Create individual columns for Year, Date and Day to check if we can explore even deeper patterns
(Since the above patterns are completely rough)

```python
data['date_year'] = data['date_parsed'].dt.year
```

```python
data['date_month'] = data['date_parsed'].dt.month
```

```python
data['date_day'] = data['date_parsed'].dt.day
```
