```python
csv_data = spx_data[['Date','(R1) Open','(R1) High','(R1) Low','(R1) Close']]
csv_data
```

```python
csv_data = csv_data.rename(columns={'Date': 'date', '(R1) Open': 'open', '(R1) High': 'high', '(R1) Low': 'low', '(R1) Close': 'close'})
```
