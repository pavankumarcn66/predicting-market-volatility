## Since, there are lot of duplicate values. Let's find the mean value for the polarity of each unique date.

    i. Group by the date
    ii. Find the mean polarity value for each date

```python
unique_date = pd.unique(json_data['date_parsed'])
unique_date 
```

```python
json_data
```

```python
json_data = json_data.groupby(['date_parsed']).mean().reset_index()
json_data
```

```python
 def analysis(score):
    if score < 0:
        return 'Negative'
    elif score > 0:
        return 'Positive'
    else:
        return 'Neutral'
```

```python
json_data['cont_sent'] = json_data.apply(lambda x: analysis(x['cont_pol']),axis =1)
json_data['cont_sent'] 
```
